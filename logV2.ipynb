```python
import numpy as np
import pandas as pd
```

```python
log=pd.read_csv("simulation_simulation_sugarAlcohols_001_OUTPUT_d-sorbitol_193l_16-1_60A_v2.log", sep="\t")
```

```python
log.info()
```

```python
log.describe()
```

```python
N = 1000
mylist=log['Total Energy (kJ/mole)']
cumsum, moving_aves = [0], []

for i, x in enumerate(mylist, 1):
    cumsum.append(cumsum[i-1] + x)
    if i>=N:
        moving_ave = (cumsum[i] - cumsum[i-N])/N
        #can do stuff with moving_ave here
        moving_aves.append(moving_ave)
```

```python
import matplotlib.pyplot as plt
```

```python
l = plt.plot(log['Time (ps)'][999:], moving_aves, 'ro')
plt.setp(l, markersize=1)
plt.setp(l, markerfacecolor='C0')

plt.show()
```
